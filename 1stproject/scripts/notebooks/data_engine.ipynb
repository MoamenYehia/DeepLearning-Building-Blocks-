```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset

```

```python
x_raw=torch.randn(100,4) # 100 samples , 4 features 

y_raw=torch.randint(0,3,(100,)) # 100 samples , 3 classes

dataset=TensorDataset(x_raw,y_raw)

data=DataLoader(dataset,batch_size=16,shuffle=True)


```

```python
for batch_X, batch_y in data:
    print("Batch X shape:", batch_X.shape)
    print("Batch y shape:", batch_y.shape)
    break
```

```python
model=nn.Sequential(
    nn.Linear(in_features=4,out_features=8),
    nn.ReLU(),
    nn.Linear(in_features=8,out_features=3)
)

print(model)
```

```python
test_output=model(batch_X) # Prediction 
print("Test output shape:", test_output.shape)
```

```python
criterion=nn.CrossEntropyLoss() # Initializing the loss function

optimizer=torch.optim.SGD(model.parameters(),lr=0.01) # Initializing the optimizer 
```

```python
num_epochs=5

for epoch in range(num_epochs):

    epoch_loss = 0.0

    for batch_X, batch_y in data:
        
     model.zero_grad() # Zero the gradients before backpropagation

     predictions=model(batch_X) # Forward pass

     loss=criterion(predictions,batch_y) # Calculate the loss

     loss.backward() # Backpropagation

     optimizer.step() # Update the parameters


epoch_loss += loss.item()

average_loss = epoch_loss / len(data)
print(f"Epoch: {epoch} | Average Loss: {average_loss:.4f}")
```
